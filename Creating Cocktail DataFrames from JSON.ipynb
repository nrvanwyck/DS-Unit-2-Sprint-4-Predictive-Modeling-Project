```python
import json

with open('master_recipes.json') as json_file:
    master_recipes = json.load(json_file)

for edition in master_recipes:
    print(len(master_recipes[edition]), '\t', edition)
```

```python
ed_1935 = 'old-mr-boston-deluxe-official-bartenders-guide-1935'
ed_1957 = 'old-mr-boston-de-luxe-official-bartenders-guide-1957'
ed_1960 = 'old-mr-boston-de-luxe-official-bartenders-guide-1960'
ed_1964 = 'old-mr-boston-de-luxe-official-bartenders-guide-1964'
ed_1972 = 'old-mr-boston-de-luxe-official-bartenders-guide-1972'
ed_1974 = 'mr-boston-deluxe-official-bartenders-guide-1974'
ed_1978 = 'mr-boston-deluxe-official-bartenders-guide-1978'
ed_1984 = 'mr-boston-official-bartenders-guide-50th-anniversary-edition-1984'
ed_1988 = 'mr-boston-official-bartenders-guide-1988'
ed_2000 = 'mr-boston-official-bartenders-and-party-guide-2000-1'
ed_2012 = 'mr-boston-official-bartenders-guide-2012'

editions = [ed_1935, ed_1957, ed_1960, ed_1964, ed_1972, ed_1974, ed_1978,
            ed_1984, ed_1988, ed_2000, ed_2012]
years = [1935, 1957, 1960, 1964, 1972, 1974, 1978, 1984, 1988, 2000, 2012]
edition_dict = {years[i]: editions[i] for i in range(len(years))}

cocktails_df = pd.DataFrame()
for year in years:
    cocktails_year_df = pd.DataFrame(master_recipes[edition_dict[year]])
    cocktails_year_df['edition_year'] = year
    cocktails_df = cocktails_df.append(cocktails_year_df)

cocktails_df = cocktails_df.drop(columns=['book_id',
                                          'created_at',
                                          'updated_at',
                                          'deleted_at',
                                          'slug',
                                          'video_url',
                                          'images',
                                          'has_images',
                                          'reviewsCount',
                                          'reviews_count_relation',
                                          'difficulty_level_id',
                                          'recipe_category',
                                          'recipe_category_id',
                                          'era_id',
                                          'origination_id',
                                          'century_id',
                                          'colors',
                                          'brands',
                                          'attributes',
                                          'recipe_status_id',
                                          'glass',
                                          'recipe_ingredient'
                                         ]).reset_index(drop=True)

cocktails_df
```

```python
ingredients_list = []
for edition in editions:
    for recipe in master_recipes[edition]:
        recipe_data = []
        for ingredient in recipe['recipe_ingredient']:
            recipe_data.append(ingredient['ingredient_id'])
            recipe_data.append(ingredient['recipe_ingredient_type_id'])
            recipe_data.append(ingredient['measurement_id'])
            recipe_data.append(ingredient['amount'])
        ingredients_list.append(recipe_data)

ingredient_id_cols = ['ingredient_id_' + str(i) for i in range(14)]
recipe_ingredient_type_id_cols = ['recipe_ingredient_type_id_' + str(i) for i in range(14)]
measurement_id_cols = ['measurement_id_' + str(i) for i in range(14)]
amount_cols = ['amount_' + str(i) for i in range(14)]

ingredient_cols = []
for i in range(14):
    ingredient_cols.append(ingredient_id_cols[i])
    ingredient_cols.append(recipe_ingredient_type_id_cols[i])
    ingredient_cols.append(measurement_id_cols[i])
    ingredient_cols.append(amount_cols[i])
ingredients_df = pd.DataFrame(ingredients_list, columns=ingredient_cols)

cocktails_df = pd.concat([cocktails_df, ingredients_df], axis=1)

cols = ['id', 'name', 'edition_year', 'is_master', 'master_recipe_id'
       ] + ingredient_cols + ['ingredient_replacement_suggestions',
                             'instructions', 'glass_id', 'notes']
cocktails_df = cocktails_df[cols]

cocktails_df
```

```python
ingredients_list = []
for edition in editions:
    for recipe in master_recipes[edition]:
        for ingredient in recipe['recipe_ingredient']:
            ingredients_list.append(ingredient['ingredient'])

ingredients = list({ingredient['id']: ingredient 
                    for ingredient in ingredients_list}.values())

ingredients_df = pd.DataFrame(ingredients).drop(columns=['created_at',
                                                'updated_at',
                                                'deleted_at',
                                                'slug',
                                                'parent_ingredient'])
ingredients_df = ingredients_df[['id',
                                'name',
                                'is_base',
                                'ingredient_category_id',
                                'parent_id',
                                'notes']]

ingredients_df
```

```python
recipe_ingredient_types_list = []
for edition in editions:
    for recipe in master_recipes[edition]:
        for ingredient in recipe['recipe_ingredient']:
            recipe_ingredient_types_list.append(
                ingredient['recipe_ingredient_type'])

recipe_ingredient_types = list({recipe_ingredient_type['id']: 
                                recipe_ingredient_type for 
                                recipe_ingredient_type in 
                                recipe_ingredient_types_list}.values())

recipe_ingredient_types_df = pd.DataFrame(recipe_ingredient_types).drop(
    columns=['created_at', 'updated_at'])

recipe_ingredient_types_df
```

```python
glasses_list = []
for edition in editions:
    for recipe in master_recipes[edition]:
        glasses_list.append(recipe['glass'])
glasses_list = [i for i in glasses_list if i]

glasses = list({glass['id']: glass for glass in glasses_list}.values())

glasses_df = pd.DataFrame(glasses).drop(columns=['created_at',
                                                 'updated_at',
                                                 'deleted_at',
                                                 'slug',
                                                 'shop_url'])
                          
glasses_df = glasses_df[['id',
                         'name',
                         'capacity',
                         'notes']]

glasses_df
```
